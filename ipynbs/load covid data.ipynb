```python
dataset_dir = '/data/datasets/covid_ct/'
```

```python
image_paths = os.listdir(dataset_dir+'training_data')
```

```python
data = {'covid':[],'pneumonia':[],'normal':[]}
```

```python
for fname in os.listdir(os.path.join(dataset_dir,'training_data')):
    fullpath = os.path.join(dataset_dir,'training_data',fname)
    data[fname.split('_')[0]].append(fullpath)
```

```python
for key in data.keys():
    print(key)
    print(data[key][:1])
    print(data[key][-1:])
    print(len(data[key]))
```

```python
from PIL import Image
```

```python
images = {'covid':[],'pneumonia':[],'normal':[]}
for key in data.keys():
    print(f'loading {key} imgs')
    for img_fname in data[key]:
        images[key].append(Image.open(img_fname))
```
